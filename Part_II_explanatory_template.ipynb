```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv("Clean_data.csv")
```

## Investors Vs Loan amount Scatter Diagram

> we will plot a scatter diagram to see the relation between # of investors and the loan amount to witness the correlation

```python
sample = df.sample(45)
sample = sample[sample['Investors'] > 1]
plt.scatter(data = sample , x = 'Investors' , y = 'LoanOriginalAmount' );
plt.xlabel('investors');
plt.ylabel('loan amount');
```

we excluded the solo loans to see the effect of investors , the plot shows a positive correlation by taking a 45 row as a sample from the original data , 45 smaple statisticly can represent the original data based on t student distribution

## Locating most of Loan Amount in Categories Violin Diagram

> we plotted a violin plot to compare the categories with the loan amount and split by loan status

```python
sns.violinplot(data=df, x='ListingCategory (numeric)', y='LoanOriginalAmount', hue='LoanStatus');
```

If we look closer to the center of the violin , we can see that business has higher stomach than the others and bigger node around 15k loans

## Clients' Monthly Income's Range by Histogram Diagram

>we created a histogram to see the clients' monthly income's range , to focus on them

```python
threshold = df['StatedMonthlyIncome'].mean() + df['StatedMonthlyIncome'].std()*4;
plt.hist(data=df, x='StatedMonthlyIncome', bins=700);
plt.xlim(0,threshold);
plt.xlabel('StatedMonthlyIncome');
plt.ylabel('count');
```

As we can clearly see , from 5k to 10k is the range of most of our clients , knowing that we can create offers to target them
